<H1> <B> Handling Missing Data </B> </H1>

```python
df['director'].fillna("Unknown", inplace=True)
```

```python
df['cast'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum() #identify missing values

```

```python
df['country'].fillna('Unknown',inplace=True)
df['date_added'].fillna('Unknown',inplace=True)
df['rating'].fillna('Unknown',inplace=True)
df['duration'].fillna(0,inplace=True)
```

```python
df = df.drop_duplicates(subset=['show_id', 'title',])
```

<H1> <B> Date Processing</B> </H1>

```python
#convert date added to dateitme format
df['date_added'] = df['date_added'].str.strip() #strip removes white space
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors ='coerce')
```

<H1> <B> Category Standardization</B> </H1>

```python
#standarizing categories:
rating_categories = {
    'TV-MA': 'Adult', 'R': 'Adult', 'PG-13': 'Family', 'TV-14': 'Family',
    'TV-PG': 'Family', 'PG': 'Children', 'G': 'Children', 'TV-Y': 'Children'
}
df['rating_category'] = df['rating'].map(rating_categories)
```

<H1> <B> Outlier checking </B> </H1>

```python
Q1 = df['popularity'].quantile(.25)
Q3 = df['popularity'].quantile(.75)
IQR = Q3- Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 *IQR
```

```python
outliers = df[(df['popularity'] < lower_bound) | (df['popularity'] > upper_bound)]
print(f'Number of outliers: {len(outliers)}')
```

```python
# Split and count genres from the 'listed_in' column
genre_counts = df['listed_in'].str.split(', ').explode().value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(y=genre_counts.index, x=genre_counts.values, orient='h')
plt.title("Distribution of Content Genres (Listed In)")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
# Split the 'listed_in' column and create binary columns for each genre
df_genres = df['listed_in'].str.get_dummies(sep=', ')

# Concatenate the new genre columns with the original dataframe
df = pd.concat([df, df_genres], axis=1)

# Display the first few rows to check the result
df.head()
```

```python
# Select numerical columns for correlation analysis
numerical_df = df[['popularity', 'average_age_watched', 'gender_male']]

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
